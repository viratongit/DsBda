```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('fb.csv')
```

```python
df
```

```python
#CREATE SUB SET
subset1=df[['like','share','comment']]
subset1
```

```python
subset2=df[['paid','posthour','postweekday']]
subset2
```

```python
#MERGE DATA
merged=pd.merge(subset1,subset2,left_on='like',right_on='paid')
merged
```

```python
# Sorting merged_data in descending order wrt 'Like'
merged.sort_values(by=['like'],ascending=False)
```

```python
# Transposing Data
merged.transpose()
```
