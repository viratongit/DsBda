```python
#Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for assignment no. 2 and 3 ( Group B)
#import dependencies
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('heartdisease.csv')
```

```python
df.head(5)
```

```python
#Barplot using seaborn
sns.barplot(x='sex', y='thalach', data = df)
```

```python
#ScatterPlot using seaborn
sns.scatterplot(x='age', y='chol', data = df)
```

```python
#Lineplot using seaborn
sns.lineplot(x='chol', y='age', data = df)
```

```python
#ScatterPlot using Matplotlib
plt.scatter(df['thalach'], df['age'])
plt.show()
```

```python
#PiePlot using Matplotlib
sex_df = pd.DataFrame(df['sex'])
sex_df
```

```python
plt.figure(figsize=(6, 6))
plt.pie(sex_df['sex'].value_counts(), labels=sex_df['sex'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Sex')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plot bar chart
plt.subplot(1, 2, 2)  # Subplot with 1 row, 2 columns, position 2
plt.bar(sex_df['sex'].value_counts().index, sex_df['sex'].value_counts())
plt.title('Bar Plot - Distribution of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```
